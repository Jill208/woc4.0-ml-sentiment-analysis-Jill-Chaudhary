# Extracting Numbers from string:


```python
car['Unique ID'] = car['Unique ID'].str.extract('(\d+)').astype(int)
```

```python
car
```

# Label Encoding:


```python
from sklearn import preprocessing
encoder  = preprocessing.LabelEncoder()
s=car.columns
for i in s:
    if car[i].dtype=='object':
        car[i]=encoder.fit_transform(car[i])
car.head()
```

# Dealing with NaN Values: 

```python
car.isna().sum()
```

```python
car=car[car['Car'].notna()]
```

```python
car.isna().sum()
```

```python
car=car[car['BuildingArea'].notna()]
```

```python
car.isna().sum()
```

```python
car=car[car['YearBuilt'].notna()]
car.isna().sum()
```

```python
car=car[car['CouncilArea'].notna()]
car.isna().sum()
```
